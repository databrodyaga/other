Will try to make simple byte-pair tokenizer

```python
# make dict with byte-tokens
vocab = {i:i for i in range(256)}
```

```python
with open("the-verdict.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()
print("total number of characters:", len(raw_text))
print(raw_text[:99])
print()

raw_data = list(raw_text.encode("utf-8"))
raw_data[:10]
```

```python
def find_bp(raw_data):

    dict_tmp = {}
    for i in range(len(raw_data)-1):
        a,b = raw_data[i],raw_data[i+1]
        if (a,b) in dict_tmp:
            dict_tmp[(a,b)]+=1
        else:
            dict_tmp[(a,b)] = 1
            
    bp = max(dict_tmp.items(), key= lambda x: x[1]) # use max for every dict_tmp.value
    print(bp)
    return bp[0]

print(find_bp(raw_data))
```

```python
while len(vocab) <= 350: #tokens number
    bp = find_bp(raw_data)
    new_token = max(vocab.values())+1
    new_raw_data=[]
    i=0
    while i <= len(raw_data)-2:
        a,b = raw_data[i],raw_data[i+1]
        if (a,b) == bp:
            new_raw_data.append(new_token)
            i+=1
        else:
            new_raw_data.append(raw_data[i])
        i+=1
    raw_data = new_raw_data.copy()
    vocab[bp]=new_token

vocab
```

```python
def vocab_encode(vocab, text):
    text = text.encode("utf-8")
    vocab = list(vocab.items())[256:]
    
    for token in vocab:
        
        new_text=[]
        i=0
        while i < len(text)-1:
            a,b = text[i],text[i+1]
            if (a,b) == token[0]:
                new_text.append(token[1])
                i+=1
            else:
                new_text.append(text[i])
            i+=1
        if i == len(text)-1: #add last byte
            new_text.append(text[i])
        text = new_text.copy()
    print('Tokenized length after encode:', len(text))
    return(text)


def vocab_decode(vocab, text):
    print('Tokenized length for decode:', len(text))
    vocab = list(vocab.items())[256:]

    for token in vocab[::-1]:
        
        new_text=[]
        i=0
        while i < len(text):
            if text[i] == token[1]:
                new_text.extend(token[0])
            else:
                new_text.append(text[i])
            i+=1

        text = new_text.copy()
    
    text = bytes(text).decode("utf-8")     
    return(text)
```

Encoding of new text:
- encode new text into bytes with text.encode("utf-8")
- iterate through vocab from id 256 (merge rules) to the last one
- apply each rule across the whole text (the order is important)
- in the end we get tokens: single bytes + merged bytes and that's our tokenized text

```python
print("Raw text length before encoding:",len(raw_text))
encode_text = vocab_encode(vocab,raw_text)    
print()
decode_text = vocab_decode(vocab,encode_text)
print("Raw text length after decoding:",len(decode_text))

raw_text == decode_text
```
