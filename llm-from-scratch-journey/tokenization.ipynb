```python
with open("chehov.txt", "r", encoding="utf-8") as f:
    raw_text = f.read()
print("total number of characters:", len(raw_text))
print(raw_text[:99])
```

```python
import re
text = "Hello, world. This, is a test."
result = re.split(r'(\s)', text)
print(result)
```

```python
result = re.split(r'([,.]|\s)', text)
print(result)
```

```python
result = [item.strip() for item in result if item.strip()]
print(result)
```

```python
#add all other dividers and filter out the spaces
text = "Helo, world. Is this-- a test?"
result = re.split(r'([,.:;?_!"()\']|--|\s)', text)
result = [item for item in result if item.strip()]
print(result)
```
